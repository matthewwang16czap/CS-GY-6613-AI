Bounding Box Prediction: predict bounding boxes of objects, using any CNN model in transfer learning.

Kalman Filter with linear velocity: Each state has parameters and their velocity. The state $[x, y, a, h, vx, vy, va, vh]$ is represented by 2 vectors which are mean and covariance. If using $K_t$ as the Kalman gain, $P^−_t$ as the covariance matrix before the measurement, $H_t$ as the measurement model, $R_t$ as measurement noise covariance matrix, $s^-_t$ as the state estimate before the measurement, $s^+_t$ as the state estimate after the measurement, $z_t$ as the measurement, the Kalman gain is:
    $$K_t = P^−_t H_t^T (H_t P^−_t H_t^T + R_t)^{-1}$$
The update state is:
    $$s^+_t = s^-_t + K_t (z_t - H_t s^-_t)$$
    
Hungarian Algorithm: Assign Bounding Box Predictions to track states with Manhalanobis distance and Deep Appearance Descriptor in $O(n^3)$. Final IoU distance gives the score for assignment.

Deep Appearance Descriptor: a KNN algorithm to detect object appearance with CNN to extracte features. Enhence Hungarian Algorithm to better assignment.

Cascade Matching: an ordering assignment to prefer existing detected object track more. Enhence Hungarian Algorithm to better assignment.
